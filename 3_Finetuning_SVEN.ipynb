```python
train_dataset = Dataset.from_pandas(train_df.reset_index(drop=True))
val_dataset = Dataset.from_pandas(val_df.reset_index(drop=True))
test_dataset = Dataset.from_pandas(test_df.reset_index(drop=True))
```

```python
dataset_dict = DatasetDict({
    "train": train_dataset,
    "validation": val_dataset,
    "test": test_dataset
})
```

```python
print(dataset_dict)
```

# Load Model

```python
model_name = "facebook/nllb-200-distilled-600M"
tokenizer = AutoTokenizer.from_pretrained(model_name, use_auth_token=token)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name, use_auth_token=token).to("cuda")
```

```python
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
```

# Preprocess Dataset

```python
src_lang = "swe_Latn"
tgt_lang = "eng_Latn"
```

```python
def preprocess(example):
    # SV and EN: column names in dataset
    src = example.get("SV", "")
    tgt = example.get("EN", "")

    # Tokenise Input
    tokenizer.src_lang = src_lang
    model_inputs = tokenizer(
        src,
        max_length=128,
        truncation=True,
        padding="max_length"
    )

    # Tokenise Target
    tokenizer.src_lang = tgt_lang
    labels = tokenizer(
        tgt,
        max_length=128,
        truncation=True,
        padding="max_length"
    )

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# Preprocess with batching
tokenized_dataset = dataset_dict.map(preprocess, batched=True)
```

# Fine-Tuning SVEN 1e-4

```python
training_args = Seq2SeqTrainingArguments(
    output_dir="/content/drive/MyDrive/Masterarbeit/Models/nllb-sv-en-1e-4-10",
    learning_rate=1e-4,
    num_train_epochs=10,
    per_device_train_batch_size=4,
    gradient_accumulation_steps=2,
    fp16=torch.cuda.is_available(),
    eval_strategy="epoch",
    save_strategy="epoch",
)

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["validation"],
    tokenizer=tokenizer,
    data_collator=data_collator,
)
```

```python
trainer.train()
```

# Generation of Translations for Evaluation

## Translations With 10th Epoch Model (last epoch) @lr 1e-4

```python
tokenizer = AutoTokenizer.from_pretrained("/content/drive/MyDrive/Masterarbeit/Models/nllb-sv-en-1e-4-10/checkpoint-1350")
model = AutoModelForSeq2SeqLM.from_pretrained("/content/drive/MyDrive/Masterarbeit/Models/nllb-sv-en-1e-4-10/checkpoint-1350")

tokenizer.src_lang = src_lang
forced_bos_token_id = tokenizer.convert_tokens_to_ids(tgt_lang)
```

```python
model.eval()
model.to("cuda")
```

```python
test_original = dataset_dict["test"].to_pandas()
test_data = tokenized_dataset["test"]

mt_output = []

for example in tqdm(test_data, desc="Translating"):
    input_ids = torch.tensor(example["input_ids"]).unsqueeze(0).to(device)
    attention_mask = torch.tensor(example["attention_mask"]).unsqueeze(0).to(device)

    with torch.no_grad():
        generated_tokens = model.generate(
            input_ids=input_ids,
            attention_mask=attention_mask,
            forced_bos_token_id=forced_bos_token_id,
            num_beams=5,
            max_length=256
        )

    translation = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
    mt_output.append(translation[0])


#add to df
test_original.insert(0, "EN model translation", mt_output)

output_path = "/content/drive/MyDrive/Masterarbeit/Models/Outputs/eval_SVEN_1e-4_ep10_last.csv"
test_original.to_csv(output_path, index=False)
```
